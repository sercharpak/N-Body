```python
class Simulation_Parallel:
    def __init__(self, N, N_t, N_p, t_t, t_f, t_g, t_total):
        self.N = N
        self.N_t=N_t
        self.N_p = N_p
        self.timings = {"tree": t_t, "force": t_f, "gather": t_g,"total":t_total}
```
